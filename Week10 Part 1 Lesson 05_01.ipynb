```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
from mpl_toolkits import mplot3d
from matplotlib import cm
```

```python
def plot_3d(x, y, u, name):
    fig = pyplot.figure(figsize=(8.0, 6.0))
    ax = mplot3d.Axes3D(fig)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('Temperature')
    X, Y = numpy.meshgrid(x, y)
    temp = numpy.random.randint(4)
    temp_Arr = [cm.viridis, cm.plasma, cm.inferno, cm.magma, cm.cividis]
    ax.plot_surface(X, Y, u, cmap=temp_Arr[temp])
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(y[0], y[-1])
    ax.view_init(elev=30.0, azim=45.0)
    ax.set_title(name)
    return
```

```python
def analyical(x, y, Lx, Ly):
    # convert two 1D array into two 2D array
    X, Y = numpy.meshgrid(x, y)
    u = numpy.sinh(1.5 * numpy.pi * Y / Ly) \
        / numpy.sinh(1.5 * numpy.pi * Lx / Ly) \
        * numpy.sin(1.5 * numpy.pi * X / Lx)
    return u
```

```python
Lx, Ly = 1.0, 1.0
nx, ny = 41, 41
x = numpy.linspace(0.0, Lx, num=nx)
y = numpy.linspace(0.0, Ly, num=ny)

ua = analyical(x, y, Lx, Ly)
plot_3d(x, y, ua, name='analytical')
```

```python
def relative_l2_norm(u, u_ref):
    rdiff = numpy.sqrt(numpy.sum((u - u_ref)**2)) \
            / numpy.sqrt(numpy.sum(u_ref**2))
    return rdiff
```

```python
def laplacian2d_jacobi(u0, rtol=1e-6, maxiter=20000):
    u = u0.copy()
    k = 0
    rdiff = rtol + 1.0
    while rdiff > rtol and k < maxiter:
        u_last = u.copy()
        u[1:-1, 1:-1] = 0.25 * (u[1:-1, :-2] + u[1:-1, 2:]
                               +u[:-2, 1:-1] + u[2:, 1:-1])
        # BC at x = Lx
        u[1:-1, -1] = u[1:-1, -2]
        
        # BC at x = 0.0
        u[1:-1, 0] = 0.0
        
        # BC at y = 0.0
        u[0, 1:-1] = 0.0
        
        rdiff = relative_l2_norm(u, u_last)
        k += 1
    return u, k, rdiff
```

```python
# initial values
u0 = numpy.zeros((ny, nx))
u0[:, 0] = 0.0
u0[0, :] = 0.0
u0[-1, :] = numpy.sin(1.5 * numpy.pi * x / Lx)

plot_3d(x, y, u0, name='initial condition')
```

```python
u, k, rdiff = laplacian2d_jacobi(u0, rtol=1e-8)
print(rdiff)
plot_3d(x, y, u, name='numerical')
```

```python
nx_values = [11, 21, 41, 81]
errors = []
for nx in nx_values:
    x = numpy.linspace(0.0, Lx, num=nx)
    y = x.copy()
    u0 = numpy.zeros((nx, nx))
    u0[-1, :] = numpy.sin(1.5 * numpy.pi * x / Lx)
    u = laplacian2d_jacobi(u0, rtol=1e-8)[0]
    ua = analyical(x, y, Lx, Ly)
    error = relative_l2_norm(u, ua)
    errors.append(error)
print(errors)
pyplot.plot(nx_values, errors, marker='o');
```

```python
pyplot.figure(figsize=(6.0, 6.0))
pyplot.grid()
dx_values = []
for nx in nx_values:
    dx_values.append(Lx / (nx - 1))
pyplot.loglog(dx_values, errors, marker='o');
pyplot.axis('equal');
```

### It is not 2nd order convergence
The boundary condition for Neumann condition is 1st-order method approximation.

So we can use central difference to analysis Neumann conditon.

```python
def w_right_BC_laplacian2d_jacobi(u0, rtol=1e-6, maxiter=20000):
    u = u0.copy()
    k = 0
    rdiff = rtol + 1.0
    while rdiff > rtol and k < maxiter:
        u_last = u.copy()
        u[1:-1, 1:-1] = 0.25 * (u[1:-1, :-2] + u[1:-1, 2:]
                               +u[:-2, 1:-1] + u[2:, 1:-1])
        # BC at x = Lx
        # u[1:-1, -1] = u[1:-1, -2]
        u[1:-1, -1] = 0.25 * (u[:-2, -1] + 2 * u[1:-1, -2] + u[2:, -1])
        
        # BC at x = 0.0
        u[1:-1, 0] = 0.0
        
        # BC at y = 0.0
        u[0, 1:-1] = 0.0
        
        rdiff = relative_l2_norm(u, u_last)
        k += 1
    return u, k, rdiff
```

```python
nx_values = [11, 21, 41, 81]
errors = []
for nx in nx_values:
    x = numpy.linspace(0.0, Lx, num=nx)
    y = x.copy()
    u0 = numpy.zeros((nx, nx))
    u0[-1, :] = numpy.sin(1.5 * numpy.pi * x / Lx)
    u = w_right_BC_laplacian2d_jacobi(u0, rtol=1e-8)[0]
    ua = analyical(x, y, Lx, Ly)
    error = relative_l2_norm(u, ua)
    errors.append(error)
print(errors)
pyplot.plot(nx_values, errors, marker='o');
```

```python
pyplot.figure(figsize=(6.0, 6.0))
pyplot.grid()
dx_values = []
for nx in nx_values:
    dx_values.append(Lx / (nx - 1))
pyplot.loglog(dx_values, errors, marker='o');
pyplot.axis('equal');
```
