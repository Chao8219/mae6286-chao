### Posion Equation

```python
import numpy
from matplotlib import pyplot
%matplotlib inline

pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16

from mpl_toolkits import mplot3d
from matplotlib import cm
```

```python
def plot_3d(x, y, u, name):
    fig = pyplot.figure(figsize=(8.0, 6.0))
    ax = mplot3d.Axes3D(fig)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('Temperature')
    X, Y = numpy.meshgrid(x, y)
    temp = numpy.random.randint(4)
    temp_Arr = [cm.viridis, cm.plasma, cm.inferno, cm.magma, cm.cividis]
    ax.plot_surface(X, Y, u, cmap=temp_Arr[temp])
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(y[0], y[-1])
    ax.view_init(elev=30.0, azim=45.0)
    ax.set_title(name)
    return
```

```python
def poisson_bound(x, y, Lx, Ly):
    X, Y = numpy.meshgrid(x, y)
    b = - 2.0 * ((numpy.pi / Lx) * (numpy.pi / Ly)) * numpy.sin(numpy.pi * X / Lx) * numpy.cos(numpy.pi * Y / Ly)
    return b
```

```python
def relative_l2_norm(u, u_ref):
    rdiff = numpy.sqrt(numpy.sum((u - u_ref)**2)) \
            / numpy.sqrt(numpy.sum(u_ref**2))
    return rdiff
```

```python
def possion2d_jacobi(u0, dx, dy, b0, rtol=1e-6, maxiter=20000):
    u = u0.copy()
    k = 0
    conv = []
    rdiff = rtol + 1.0
    b = b0.copy()
    while rdiff > rtol and k < maxiter:
        u_last = u.copy()
        u_i_plus_1 = u[1:-1, 2:]
        u_i_mins_1 = u[1:-1, :-2]
        u_j_plus_1 = u[2:, 1:-1]
        u_j_mins_1 = u[:-2, 1:-1]
        u[1:-1, 1:-1] = ((u_i_plus_1 + u_i_mins_1) * (dy**2) \
                        + (u_j_plus_1 + u_j_mins_1) * (dx**2) \
                        - b[1:-1, 1:-1] * (dx**2) * (dy**2)) \
                        / (2 * (dx**2 + dy**2))
        rdiff = relative_l2_norm(u, u_last)
        k += 1
        conv.append(rdiff)
    return u, k, conv
```

```python
Lx, Ly = 1.0, 1.0
Lx_begin, Lx_end = 0.0, 1.0
Ly_begin, Ly_end = -0.5, 0.5
nx, ny = 41, 41
x = numpy.linspace(Lx_begin, Lx_end, num=nx)
y = numpy.linspace(Ly_begin, Ly_end, num=ny)
dx = Lx / (nx - 1)
dy = Ly / (ny - 1)
```

```python
u0 = numpy.zeros((ny, nx))
b = poisson_bound(x, y, Lx, Ly)
plot_3d(x, y, u0, name='Initial Condition')
```

```python
u, ite, conv = possion2d_jacobi(u0, dx, dy, b, rtol=2e-7)
plot_3d(x, y, u, name='Numerical')
```

```python
pyplot.semilogy(conv)
pyplot.title("convergence changes\n w.r.t. iteration times")
pyplot.xlabel("iteration times")
pyplot.ylabel("convergence")
pyplot.grid();
```

```python
def possion_anaylytical(x, y, Lx, Ly):
    X, Y = numpy.meshgrid(x, y)
    u = numpy.sin(X * numpy.pi / Lx) * numpy.cos(Y * numpy.pi / Ly)
    return u
```

```python
ua = possion_anaylytical(x, y, Lx, Ly)
plot_3d(x, y, ua, name='Analytical')
```

```python
nx_values = [11, 21, 41, 81]
errors = []
for nx in nx_values:
    ny = nx
    dx = Lx / (nx - 1)
    dy = Ly / (ny - 1)
    x = numpy.linspace(Lx_begin, Lx_end, num=nx)
    y = numpy.linspace(Ly_begin, Ly_end, num=ny)
    u0 = numpy.zeros((ny, nx))
    b = poisson_bound(x, y, Lx, Ly)
    u, ite, conv = possion2d_jacobi(u0, dx, dy, b, rtol=2e-7)
    u_exact = possion_anaylytical(x, y, Lx, Ly)
    errors.append(relative_l2_norm(u, u_exact))
```

```python
dx_values = []
for nx in nx_values:
    dx_values.append(Lx / (nx - 1))
pyplot.figure(figsize=(5.0, 5.0))
pyplot.loglog(dx_values, errors)
pyplot.grid()
pyplot.axis('equal');
```
